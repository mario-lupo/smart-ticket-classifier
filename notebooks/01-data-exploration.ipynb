```python
print("Abbiamo visto, attraverso df.info, che ci sono vari dati mancanti.")
print("con isnull() otteniamo un dataframe di booleani True/False per ogni cella del dataframe originale, dove True indica un valore nullo.")
print("ma per calcolare la percentuale di valori nulli per colonna, dobbiamo prima sommare i valori True (che sono considerati come 1) per ogni colonna con sum(), ottenendo cos√¨ il numero di valori nulli per colonna")
print("Usiamo df.isnull().sum() per contare quanti valori nulli ci sono in ogni colonna, e poi calcoliamo la percentuale di valori nulli per colonna.")
print("poi dividiamo per il numero totale di righe (len(df)) (quali righe? Tutte le righe di quella colonna, quindi se ho 50 valori nulli su 1000 righe in quella colonna , 50/1000 = 0.05)")
null_pct = (df.isnull().sum())
print(f"\nPercentuale di valori nulli per colonna:\n{null_pct}\n\n")
for i in null_pct.index:
    if null_pct[i] > 0:
        print(f"La colonna '{i}' ha {null_pct[i]} valori nulli, che rappresentano il {null_pct[i]/len(df)*100:.2f}% del totale.")
# perch√© ho fatto .index? Perch√© null_pct √® una serie pandas, e lo uso per iterare sulle sue etichette (i nomi delle colonne)
# infatti .index mi restituisce un oggetto Index con i nomi delle colonne della serie
# quindi in questo caso sar√† equivalente a fare for i in null_pct.keys()
```
